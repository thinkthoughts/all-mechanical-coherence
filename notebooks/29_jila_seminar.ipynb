## 1. Clone repo, enter repo root, and install dependencies

````python
from pathlib import Path
import os
import subprocess
import sys

REPO = "all-mechanical-coherence"
GITHUB_USER = "thinkthoughts"
BRANCH = "main"

CONTENT = Path("/content")
ROOT = CONTENT / REPO

if not ROOT.exists():
    url = f"https://github.com/{GITHUB_USER}/{REPO}.git"
    print(f"Cloning {url}")
    subprocess.run(["git", "clone", "--branch", BRANCH, url, str(ROOT)], check=True)
else:
    print(f"Repo already exists: {ROOT}")

os.chdir(ROOT)

subprocess.run([sys.executable, "-m", "pip", "install", "-q", "pyyaml"], check=True)

print("Working directory:", Path.cwd())
print("Repo contents:")
for item in sorted(Path.cwd().iterdir()):
    print("-", item.name)
````

## 2. Load `paper.yaml`

````python
import yaml

PAPER_YAML = ROOT / "paper.yaml"

if not PAPER_YAML.exists():
    raise FileNotFoundError(f"paper.yaml not found at {PAPER_YAML}")

with open(PAPER_YAML, "r", encoding="utf-8") as f:
    data = yaml.safe_load(f)

paper = data["paper"]
seminar = data.get("seminar", {})

print(paper["title"])
print("arXiv:", paper["arxiv"])
print("URL:", paper["url"])
print()
print("Seminar:", seminar.get("title", ""))
print("Speaker:", seminar.get("speaker", ""))
print("Date:", seminar.get("date", ""))
````

## 10. Write seminar notes file

This writes a Markdown notes file.

After the seminar, edit `seminar_notes` above and rerun this cell.

````python
reports_dir = ROOT / "reports"
reports_dir.mkdir(exist_ok=True)

lines = [
    "# 29 — JILA Seminar Notes",
    "",
    f"**Paper:** {paper['title']}",
    f"**arXiv:** {paper['arxiv']}",
    f"**Speaker:** {seminar.get('speaker', '')}",
    f"**Seminar:** {seminar.get('title', '')}",
    f"**Date:** {seminar.get('date', '')}",
    "",
    "## Pre-seminar summary",
    "",
]

for key, value in pre_seminar_summary.items():
    lines.append(f"- **{key}:** {value}")

lines.extend([
    "",
    "## Questions",
    "",
])

for i, question in enumerate(questions, start=1):
    lines.append(f"{i}. {question}")

lines.extend([
    "",
    "## Best single question",
    "",
    best_question,
    "",
    "## Live notes",
    "",
])

for key, value in seminar_notes.items():
    lines.append(f"- **{key}:** {value}")

lines.extend([
    "",
    "## Constraint tracking",
    "",
])

for row in constraint_tracking:
    lines.append(f"- **{row['constraint']}:** {row['seminar status']}")

lines.extend([
    "",
    "## Repo decision",
    "",
    f"Possible next repo: `{repo_decision['possible_next_repo']}`",
    "",
    "Extend this repo if:",
])

for item in repo_decision["extend_this_repo_if"]:
    lines.append(f"- {item}")

lines.extend([
    "",
    "Start next repo if:",
])

for item in repo_decision["start_next_repo_if"]:
    lines.append(f"- {item}")

out = reports_dir / "29_jila_seminar_notes.md"
out.write_text("\n".join(lines), encoding="utf-8")

print("Wrote:", out)
````
