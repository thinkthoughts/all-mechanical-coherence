# 29 Gate Pathways

Can all-mechanical coherence protection create a pathway toward phonon-mediated quantum gates?

## Central question

> Can coherence protection create a pathway toward phonon-mediated quantum gates?

## Working statement

The paper does not demonstrate gates. It expands the coherence and coupling regime that future gate architectures could exploit.

This notebook keeps the gate-pathway analysis compact:

1. interaction-cycle feasibility map
2. protected coherence window
3. two-spin mechanical phonon bus
4. coherence-protection-to-gate hierarchy


```python
from pathlib import Path
import zipfile

import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import pandas as pd

FIGURES = Path("figures")
FIGURES.mkdir(exist_ok=True)

RESULTS = Path("results")
RESULTS.mkdir(exist_ok=True)

plt.rcParams["figure.dpi"] = 120
```

## 1. Interaction-cycle feasibility map

The useful gate-pathway quantity is the number of interaction cycles available before protected coherence is lost.

Use a schematic proxy:

\[
\text{interaction cycles} \sim T_{coherence} g_{spin-phonon}
\]

This figure shows why coherence protection and spin–phonon coupling must improve together.


```python
coherence = np.linspace(1, 10, 300)
coupling = np.linspace(0.05, 1.0, 300)

T, G = np.meshgrid(coherence, coupling)

cycles = T * G

fig, ax = plt.subplots(figsize=(8, 6))

im = ax.imshow(
    cycles,
    origin="lower",
    aspect="auto",
    extent=[coherence.min(), coherence.max(), coupling.min(), coupling.max()],
)

contours = ax.contour(
    T,
    G,
    cycles,
    levels=[1.0, 2.5, 5.0, 7.5],
    colors="black",
    linewidths=0.8,
)
ax.clabel(contours, inline=True, fontsize=8)

ax.scatter([3.0], [0.25], s=90, label="reference regime")
ax.scatter([7.0], [0.65], s=90, label="expanded regime")

ax.text(2.4, 0.18, "single-qubit\ncontrol", ha="center", fontsize=10)
ax.text(6.6, 0.62, "entangling\noperation regime", ha="center", fontsize=10)

ax.set_title("Interaction-Cycle Feasibility Map")
ax.set_xlabel("Protected coherence time, normalized")
ax.set_ylabel("Spin–phonon coupling, normalized")
ax.legend(loc="upper left")

cbar = fig.colorbar(im, ax=ax)
cbar.set_label("Available interaction cycles")

output = FIGURES / "29_gate_feasibility_map.png"
plt.savefig(output, dpi=300, bbox_inches="tight")
print(f"saved: {output}")
plt.show()
```

## 2. Protected coherence window

The gate-pathway question is not only whether fast control exists. It is whether an interaction can fit inside a protected coherence window.

This figure avoids over-specifying gate timing and keeps the design claim simple:

\[
\text{longer protected coherence} \rightarrow \text{more available interaction time}
\]


```python
fig, ax = plt.subplots(figsize=(9, 4.8))
ax.set_xlim(0, 10)
ax.set_ylim(0, 3)
ax.axis("off")

ax.text(
    5,
    2.85,
    "Mechanical Dressing Expands the Coherence Window",
    ha="center",
    va="top",
    fontsize=17,
    fontweight="bold",
)

# Reference window
ax.text(0.6, 2.15, "reference\nregime", ha="center", va="center", fontsize=11)
ax.hlines(2.15, 1.4, 4.4, linewidth=8)
ax.plot([1.4, 1.4], [2.03, 2.27], color="black", linewidth=2)
ax.plot([4.4, 4.4], [2.03, 2.27], color="black", linewidth=2)

# Expanded window
ax.text(0.6, 1.25, "expanded\nregime", ha="center", va="center", fontsize=11)
ax.hlines(1.25, 1.4, 8.4, linewidth=8)
ax.plot([1.4, 1.4], [1.13, 1.37], color="black", linewidth=2)
ax.plot([8.4, 8.4], [1.13, 1.37], color="black", linewidth=2)

# Gate operation block inside expanded window
gate_start, gate_end = 4.9, 7.2
rect = patches.FancyBboxPatch(
    (gate_start, 0.52),
    gate_end - gate_start,
    0.46,
    boxstyle="round,pad=0.05",
    edgecolor="black",
    facecolor="white",
    linewidth=1.4,
)
ax.add_patch(rect)
ax.text((gate_start + gate_end) / 2, 0.75, "gate-feasible\ninteraction interval", ha="center", va="center", fontsize=10)

ax.annotate(
    "",
    xy=(gate_start, 1.03),
    xytext=(gate_start, 1.18),
    arrowprops=dict(arrowstyle="-|>", lw=1.2),
)
ax.annotate(
    "",
    xy=(gate_end, 1.03),
    xytext=(gate_end, 1.18),
    arrowprops=dict(arrowstyle="-|>", lw=1.2),
)

ax.text(
    5,
    0.20,
    "longer protected coherence permits more interaction cycles before decoherence dominates",
    ha="center",
    fontsize=10,
)

output = FIGURES / "29_protected_coherence_window.png"
plt.savefig(output, dpi=300, bbox_inches="tight")
print(f"saved: {output}")
plt.show()
```

## 3. Two-spin mechanical phonon bus

The mechanical mode acts as a bus when it couples two spins through a shared resonant degree of freedom.

This supports the gate-pathway concepts most relevant here:

1. information transfer
2. state exchange
3. entanglement generation


```python
fig, ax = plt.subplots(figsize=(9, 6))
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis("off")

ax.text(
    0.5,
    0.94,
    "Two-Spin Mechanical Phonon Bus",
    ha="center",
    va="top",
    fontsize=18,
    fontweight="bold",
)

bus = patches.FancyBboxPatch(
    (0.38, 0.38),
    0.24,
    0.22,
    boxstyle="round,pad=0.05",
    edgecolor="black",
    facecolor="white",
    linewidth=1.5,
)
ax.add_patch(bus)
ax.text(
    0.50,
    0.49,
    "mechanical\nmode",
    ha="center",
    va="center",
    fontsize=14,
    fontweight="bold",
)

spin_specs = [
    ("Spin A", 0.18, 0.70),
    ("Spin B", 0.82, 0.70),
]

for label, x, y in spin_specs:
    ax.text(
        x,
        y,
        label,
        ha="center",
        va="center",
        fontsize=14,
        fontweight="bold",
        bbox=dict(
            boxstyle="round,pad=0.45",
            facecolor="white",
            edgecolor="black",
            linewidth=1.4,
        ),
    )

ax.annotate(
    "",
    xy=(0.42, 0.55),
    xytext=(0.25, 0.66),
    arrowprops=dict(arrowstyle="-|>", lw=1.7),
)
ax.annotate(
    "",
    xy=(0.58, 0.55),
    xytext=(0.75, 0.66),
    arrowprops=dict(arrowstyle="-|>", lw=1.7),
)

ax.text(0.31, 0.66, "coupling", ha="center", fontsize=10)
ax.text(0.69, 0.66, "coupling", ha="center", fontsize=10)

ax.annotate(
    "",
    xy=(0.74, 0.30),
    xytext=(0.26, 0.30),
    arrowprops=dict(arrowstyle="<|-|>", lw=1.7),
)

ax.text(
    0.50,
    0.23,
    "information transfer • state exchange • entanglement generation",
    ha="center",
    fontsize=11,
)

ax.text(
    0.50,
    0.13,
    "shared mechanical mode provides the gate pathway",
    ha="center",
    fontsize=10,
)

output = FIGURES / "29_phonon_bus_concept.png"
plt.savefig(output, dpi=300, bbox_inches="tight")
print(f"saved: {output}")
plt.show()
```

## 4. From coherence protection to gate operations

This hierarchy is qualitative: the goal is to show the dependency chain rather than assign physical scale values.

\[
\text{protected coherence}
\rightarrow
\text{usable spin--phonon coupling}
\rightarrow
\text{effective spin--spin interaction}
\rightarrow
\text{gate pathway}
\]


```python
hierarchy = pd.DataFrame(
    [
        {
            "Layer": "Protected coherence",
            "Role": "available operation window",
        },
        {
            "Layer": "Spin–phonon coupling",
            "Role": "single-spin mechanical interaction",
        },
        {
            "Layer": "Effective spin–spin interaction",
            "Role": "mediated two-spin coupling",
        },
        {
            "Layer": "Gate operation",
            "Role": "controlled entangling operation",
        },
    ]
)

fig, ax = plt.subplots(figsize=(7, 7))
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis("off")

ax.text(
    0.5,
    0.95,
    "From Coherence Protection to Gate Operations",
    ha="center",
    va="top",
    fontsize=17,
    fontweight="bold",
)

levels = [
    ("Protected coherence", 0.78),
    ("Spin–phonon coupling", 0.60),
    ("Effective spin–spin interaction", 0.42),
    ("Gate operation", 0.24),
]

for label, y in levels:
    ax.text(
        0.5,
        y,
        label,
        ha="center",
        va="center",
        fontsize=13,
        fontweight="bold",
        bbox=dict(
            boxstyle="round,pad=0.45",
            facecolor="white",
            edgecolor="black",
            linewidth=1.4,
        ),
    )

for (_, y1), (_, y2) in zip(levels[:-1], levels[1:]):
    ax.annotate(
        "",
        xy=(0.5, y2 + 0.055),
        xytext=(0.5, y1 - 0.055),
        arrowprops=dict(arrowstyle="-|>", lw=1.5),
    )

ax.text(
    0.5,
    0.08,
    "protected coherence → usable spin–phonon coupling → effective spin–spin interaction → gate pathway",
    ha="center",
    fontsize=9.5,
)

output = FIGURES / "29_gate_pathway_hierarchy.png"
plt.savefig(output, dpi=300, bbox_inches="tight")
print(f"saved: {output}")
plt.show()

hierarchy
```

## 6. Interpretation

The paper does not demonstrate gates.

It expands the coherence and coupling regime that future gate architectures could exploit.

The resulting technology path is:

\[
\text{coherence protection}
\rightarrow
\text{expanded operating regime}
\rightarrow
\text{phonon-mediated interaction}
\rightarrow
\text{entanglement generation}
\rightarrow
\text{gate pathway}
\]

The key gate-level question is whether the effective interaction can fit inside the protected coherence window.
