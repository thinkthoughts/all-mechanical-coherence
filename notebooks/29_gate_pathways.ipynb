This notebook asks what engineering path connects that result to gate-level operation.

The roadmap is:

\[
\text{coherence protection}
\rightarrow
\text{usable spin--phonon coupling}
\rightarrow
\text{phonon-mediated information transfer}
\rightarrow
\text{entanglement generation}
\rightarrow
\text{gate pathways}
\]

## 2. Spin–phonon gate roadmap

A mechanical mode can mediate an interaction between spatially separated spins.

The conceptual pathway is:

\[
\text{Spin A}
\rightarrow
\text{mechanical mode}
\rightarrow
\text{Spin B}
\]


```python
fig, ax = plt.subplots(figsize=(11, 5.5))
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis("off")

ax.text(
    0.5,
    0.94,
    "Spin–Phonon Gate Roadmap",
    ha="center",
    va="top",
    fontsize=18,
    fontweight="bold",
)

nodes = [
    ("Spin A", 0.12, 0.55),
    ("Mechanical\nmode", 0.50, 0.55),
    ("Spin B", 0.88, 0.55),
]

for label, x, y in nodes:
    ax.text(
        x,
        y,
        label,
        ha="center",
        va="center",
        fontsize=14,
        fontweight="bold",
        bbox=dict(boxstyle="round,pad=0.55", facecolor="white", edgecolor="black", linewidth=1.5),
    )

ax.annotate("", xy=(0.39, 0.55), xytext=(0.22, 0.55), arrowprops=dict(arrowstyle="-|>", lw=2))
ax.annotate("", xy=(0.78, 0.55), xytext=(0.61, 0.55), arrowprops=dict(arrowstyle="-|>", lw=2))

ax.text(0.305, 0.63, "spin–phonon\ncoupling", ha="center", fontsize=10)
ax.text(0.695, 0.63, "phonon-mediated\ninteraction", ha="center", fontsize=10)

steps = [
    "mechanical dressing",
    "coherence protection",
    "spin–phonon coupling",
    "entangling interaction",
    "two-qubit gate pathway",
]

for i, step in enumerate(steps):
    x = 0.10 + i * 0.20
    ax.text(
        x,
        0.18,
        step,
        ha="center",
        va="center",
        fontsize=10,
        bbox=dict(boxstyle="round,pad=0.35", facecolor="white", edgecolor="black"),
    )
    if i < len(steps) - 1:
        ax.annotate("", xy=(x + 0.14, 0.18), xytext=(x + 0.06, 0.18), arrowprops=dict(arrowstyle="-|>", lw=1.2))

output = FIGURES / "29_spin_phonon_gate_roadmap.png"
plt.savefig(output, dpi=300, bbox_inches="tight")
print(f"saved: {output}")
plt.show()
```

## 3. Gate-pathway hierarchy

A useful gate pathway requires an ordered progression:

\[
\text{protected coherence}
\rightarrow
\text{spin--phonon coupling}
\rightarrow
\text{effective spin--spin interaction}
\rightarrow
\text{gate operation}
\]

The hierarchy is qualitative here: the goal is to show the dependency chain rather than assign physical scale values.


```python
hierarchy = pd.DataFrame(
    [
        {
            "Layer": "Protected coherence",
            "Role": "available operation window",
        },
        {
            "Layer": "Spin–phonon coupling",
            "Role": "single-spin mechanical interaction",
        },
        {
            "Layer": "Effective spin–spin interaction",
            "Role": "mediated two-spin coupling",
        },
        {
            "Layer": "Gate operation",
            "Role": "controlled entangling operation",
        },
    ]
)

fig, ax = plt.subplots(figsize=(7, 7))
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis("off")

ax.text(
    0.5,
    0.95,
    "Gate-Pathway Hierarchy",
    ha="center",
    va="top",
    fontsize=18,
    fontweight="bold",
)

levels = [
    ("Protected coherence", 0.78),
    ("Spin–phonon coupling", 0.60),
    ("Effective spin–spin interaction", 0.42),
    ("Gate operation", 0.24),
]

for label, y in levels:
    ax.text(
        0.5,
        y,
        label,
        ha="center",
        va="center",
        fontsize=13,
        fontweight="bold",
        bbox=dict(
            boxstyle="round,pad=0.45",
            facecolor="white",
            edgecolor="black",
            linewidth=1.4,
        ),
    )

for (_, y1), (_, y2) in zip(levels[:-1], levels[1:]):
    ax.annotate(
        "",
        xy=(0.5, y2 + 0.055),
        xytext=(0.5, y1 - 0.055),
        arrowprops=dict(arrowstyle="-|>", lw=1.5),
    )

ax.text(
    0.5,
    0.08,
    "coherence window → usable coupling → mediated interaction → gate pathway",
    ha="center",
    fontsize=10,
)

output = FIGURES / "29_gate_pathway_hierarchy.png"
plt.savefig(output, dpi=300, bbox_inches="tight")
print(f"saved: {output}")
plt.show()

hierarchy
```

## 4. Gate time versus coherence time

A simple gate pathway criterion is:

\[
t_{gate} < T_{coherence}
\]

For an effective coupling strength \(g_{eff}\), use:

\[
t_{gate} \sim \frac{1}{g_{eff}}
\]

The computationally useful region opens when the gate time fits inside the coherence window.


```python
g_eff = np.linspace(0.05, 1.0, 600)
t_gate = 1 / g_eff

# Normalized protected coherence windows.
T_reference = 4.0
T_expanded = 8.0

fig, ax = plt.subplots(figsize=(9, 5.5))

ax.plot(g_eff, t_gate, linewidth=2.5, label=r"gate time $\sim 1/g_{eff}$")
ax.axhline(T_reference, linestyle="--", linewidth=1.5, label="reference coherence window")
ax.axhline(T_expanded, linestyle=":", linewidth=2.0, label="expanded coherence window")

ax.fill_between(
    g_eff,
    t_gate,
    T_expanded,
    where=t_gate < T_expanded,
    alpha=0.18,
)

ax.text(0.72, 5.9, "gate time fits inside\nprotected coherence", ha="center", fontsize=10)

ax.set_title("Gate Time vs Protected Coherence")
ax.set_xlabel("Effective spin–spin coupling, normalized")
ax.set_ylabel("Time, normalized")
ax.set_ylim(0, 20)
ax.grid(True, alpha=0.3)
ax.legend()

output = FIGURES / "29_gate_times_vs_coherence.png"
plt.savefig(output, dpi=300, bbox_inches="tight")
print(f"saved: {output}")
plt.show()
```

## 5. Two-spin mechanical phonon bus concept

The mechanical mode acts as a bus when it couples two spins through a shared resonant degree of freedom.

This supports the gate-pathway concepts most relevant here:

1. information transfer
2. state exchange
3. entanglement generation


```python
fig, ax = plt.subplots(figsize=(9, 6))
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis("off")

ax.text(
    0.5,
    0.94,
    "Two-Spin Mechanical Phonon Bus",
    ha="center",
    va="top",
    fontsize=18,
    fontweight="bold",
)

# Central mechanical mode / bus
bus = patches.FancyBboxPatch(
    (0.38, 0.38),
    0.24,
    0.22,
    boxstyle="round,pad=0.05",
    edgecolor="black",
    facecolor="white",
    linewidth=1.5,
)
ax.add_patch(bus)
ax.text(
    0.50,
    0.49,
    "mechanical\nmode",
    ha="center",
    va="center",
    fontsize=14,
    fontweight="bold",
)

# Spin boxes
spin_specs = [
    ("Spin A", 0.18, 0.70),
    ("Spin B", 0.82, 0.70),
]

for label, x, y in spin_specs:
    ax.text(
        x,
        y,
        label,
        ha="center",
        va="center",
        fontsize=14,
        fontweight="bold",
        bbox=dict(
            boxstyle="round,pad=0.45",
            facecolor="white",
            edgecolor="black",
            linewidth=1.4,
        ),
    )

# Coupling arrows
ax.annotate(
    "",
    xy=(0.42, 0.55),
    xytext=(0.25, 0.66),
    arrowprops=dict(arrowstyle="-|>", lw=1.7),
)
ax.annotate(
    "",
    xy=(0.58, 0.55),
    xytext=(0.75, 0.66),
    arrowprops=dict(arrowstyle="-|>", lw=1.7),
)

ax.text(0.31, 0.66, "coupling", ha="center", fontsize=10)
ax.text(0.69, 0.66, "coupling", ha="center", fontsize=10)

# Mediated operation band
ax.annotate(
    "",
    xy=(0.74, 0.30),
    xytext=(0.26, 0.30),
    arrowprops=dict(arrowstyle="<|-|>", lw=1.7),
)

ax.text(
    0.50,
    0.23,
    "information transfer • state exchange • entanglement generation",
    ha="center",
    fontsize=11,
)

ax.text(
    0.50,
    0.13,
    "shared mechanical mode provides the gate pathway",
    ha="center",
    fontsize=10,
)

output = FIGURES / "29_phonon_bus_concept.png"
plt.savefig(output, dpi=300, bbox_inches="tight")
print(f"saved: {output}")
plt.show()
```

## 6. Gate feasibility map

The relevant axes are:

\[
T_{coherence}
\]

and

\[
g_{spin-phonon}
\]

The useful region expands when coherence protection and stronger interaction are available together.


```python
coherence = np.linspace(1, 10, 300)
coupling = np.linspace(0.05, 1.0, 300)

T, G = np.meshgrid(coherence, coupling)

# Gate feasibility proxy: more coherence and stronger coupling both help.
# The product is the number of interaction cycles available within coherence time.
cycles = T * G

fig, ax = plt.subplots(figsize=(8, 6))

im = ax.imshow(
    cycles,
    origin="lower",
    aspect="auto",
    extent=[coherence.min(), coherence.max(), coupling.min(), coupling.max()],
)

contours = ax.contour(
    T,
    G,
    cycles,
    levels=[1.0, 2.5, 5.0, 7.5],
    colors="black",
    linewidths=0.8,
)
ax.clabel(contours, inline=True, fontsize=8)

ax.scatter([3.0], [0.25], s=90, label="reference regime")
ax.scatter([7.0], [0.65], s=90, label="expanded regime")

ax.text(2.4, 0.18, "single-qubit\ncontrol", ha="center", fontsize=10)
ax.text(6.6, 0.62, "entangling\noperation regime", ha="center", fontsize=10)

ax.set_title("Gate Feasibility Map")
ax.set_xlabel("Protected coherence time, normalized")
ax.set_ylabel("Spin–phonon coupling, normalized")
ax.legend(loc="upper left")

cbar = fig.colorbar(im, ax=ax)
cbar.set_label("Available interaction cycles")

output = FIGURES / "29_gate_feasibility_map.png"
plt.savefig(output, dpi=300, bbox_inches="tight")
print(f"saved: {output}")
plt.show()
```
