# 00 — Context

Fresh-runtime notebook for `all-mechanical-coherence`.

This notebook is designed to work in Google Colab from **Runtime → Restart and run all**.

It clones the repo, installs `pyyaml`, enters the repo root, loads `paper.yaml`, and generates `reports/context_summary.md`.

## 1. Repo settings

```python
REPO = "all-mechanical-coherence"
GITHUB_USER = "thinkthoughts"
BRANCH = "main"
```

## 2. Clone repo and enter repo root

```python
from pathlib import Path
import os
import subprocess
import sys

CONTENT = Path("/content")
ROOT = CONTENT / REPO

if ROOT.exists():
    print(f"Repo already exists: {ROOT}")
else:
    url = f"https://github.com/{GITHUB_USER}/{REPO}.git"
    print(f"Cloning {url}")
    subprocess.run(["git", "clone", "--branch", BRANCH, url, str(ROOT)], check=True)

os.chdir(ROOT)

print("Working directory:", Path.cwd())
print("Repo contents:")
for item in sorted(Path.cwd().iterdir()):
    print("-", item.name)
```

## 3. Install dependencies

```python
import subprocess
import sys

subprocess.run([sys.executable, "-m", "pip", "install", "-q", "pyyaml"], check=True)

import yaml
print("pyyaml ready")
```

## 4. Load `paper.yaml`

```python
from pathlib import Path
import yaml

ROOT = Path.cwd()
PAPER_YAML = ROOT / "paper.yaml"

if not PAPER_YAML.exists():
    raise FileNotFoundError(f"Expected paper.yaml at {PAPER_YAML}")

with open(PAPER_YAML, "r", encoding="utf-8") as f:
    data = yaml.safe_load(f)

print("Loaded sections:")
for key in data.keys():
    print("-", key)
```

## 5. Paper metadata

```python
paper = data["paper"]

print(paper["title"])
print("arXiv:", paper["arxiv"])
print("Version:", paper.get("version", ""))
print("URL:", paper["url"])
print("HTML:", paper.get("html", ""))
print("Year:", paper.get("year", ""))
```

## 6. Seminar context

```python
seminar = data.get("seminar", {})

print("Event:", seminar.get("event", ""))
print("Title:", seminar.get("title", ""))
print("Speaker:", seminar.get("speaker", ""))
print("Affiliation:", seminar.get("affiliation", ""))
print("Date:", seminar.get("date", ""))
print("Context:", seminar.get("context", ""))
```

## 7. Core claim

```python
print(data["core_claim"])
```

## 8. Observations

```python
for i, observation in enumerate(data.get("observations", []), start=1):
    print(f"{i}. {observation}")
```

## 9. Engineering statements

```python
for item in data.get("engineering_statements", []):
    print(f"{item['id']} — {item['category']}")
    print(item["statement"])
    print()
```

## 10. Seminar questions

```python
for i, question in enumerate(data.get("questions_for_seminar", []), start=1):
    print(f"{i}. {question}")
```

## 11. Climate Reality framing

```python
climate = data.get("climate_reality", {})

print(climate.get("title", ""))
print()
print(climate.get("statement", ""))
print()
print("Mathematical context:", climate.get("mathematical_context", ""))
```

## 12. Toward phonon-mediated quantum gates

The present paper demonstrates all-mechanical coherence protection and fast single-qubit control. The next question is whether mechanically dressed protection can remain compatible with high-fidelity phonon-mediated two-qubit gates.

```python
gate_constraints = [
    "spin-phonon coupling strength",
    "mechanical quality factor",
    "drive stability",
    "residual magnetic noise",
    "heating",
    "cavity design",
    "optical initialization and readout",
    "two-qubit gate fidelity",
    "integration into on-chip phononic networks",
]

print("Constraints to track toward phonon-mediated quantum gates:")
for item in gate_constraints:
    print("-", item)
```

## 13. Generate `reports/context_summary.md`

```python
reports_dir = ROOT / "reports"
reports_dir.mkdir(exist_ok=True)

summary_lines = [
    f"# {data['climate_reality']['title']}",
    "",
    f"**Paper:** {paper['title']}  ",
    f"**arXiv:** {paper['arxiv']}  ",
    f"**URL:** {paper['url']}  ",
    "",
    f"**Seminar:** {seminar.get('title', '')}  ",
    f"**Speaker:** {seminar.get('speaker', '')}  ",
    f"**Date:** {seminar.get('date', '')}",
    "",
    "## Core claim",
    "",
    data["core_claim"],
    "",
    "## Observations",
    "",
]

summary_lines.extend([f"- {obs}" for obs in data.get("observations", [])])

summary_lines.extend([
    "",
    "## Engineering statements",
    "",
])

summary_lines.extend([
    f"- **{item['id']} — {item['category']}:** {item['statement']}"
    for item in data.get("engineering_statements", [])
])

summary_lines.extend([
    "",
    "## Toward phonon-mediated quantum gates",
    "",
    "The immediate result supports coherence protection and fast single-qubit control.",
    "The next research direction is to track whether mechanically dressed protection",
    "can remain compatible with high-fidelity phonon-mediated two-qubit gates.",
    "",
    "Constraints to track:",
    "",
])

summary_lines.extend([f"- {item}" for item in gate_constraints])

summary_lines.extend([
    "",
    "## Climate reality",
    "",
    data["climate_reality"]["statement"],
    "",
    data["climate_reality"].get("mathematical_context", ""),
    "",
])

summary = "\n".join(summary_lines)

out = reports_dir / "context_summary.md"
out.write_text(summary, encoding="utf-8")

print("Wrote:", out)
```
