# 00 — Context

This notebook initializes the `all-mechanical-coherence` repository.

It loads `paper.yaml`, summarizes the paper context, extracts engineering statements, and prepares the repo for the JILA seminar on all-mechanical coherence protection and fast control of a silicon-vacancy spin qubit.

The notebook is designed to run from either:

- the repository root
- the `notebooks/` directory
- Google Colab after cloning the repo

## 1. Locate the repository root and `paper.yaml`

This cell checks several likely locations so the notebook works locally and in Colab.

```python
from pathlib import Path
import sys

cwd = Path.cwd().resolve()
print("Current working directory:", cwd)

candidates = [
    cwd / "paper.yaml",
    cwd.parent / "paper.yaml",
    cwd.parent.parent / "paper.yaml",
    Path("/content/all-mechanical-coherence/paper.yaml"),
]

PAPER_YAML = None

for candidate in candidates:
    if candidate.exists():
        PAPER_YAML = candidate.resolve()
        break

if PAPER_YAML is None:
    raise FileNotFoundError(
        "Could not find paper.yaml. "
        "Run this notebook from the repo root or notebooks/ directory, "
        "or clone github.com/thinkthoughts/all-mechanical-coherence in Colab."
    )

ROOT = PAPER_YAML.parent
SRC = ROOT / "src"

if SRC.exists() and str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))

print("Repository root:", ROOT)
print("Using paper.yaml:", PAPER_YAML)
```

## 2. Load YAML metadata

```python
import yaml

with open(PAPER_YAML, "r", encoding="utf-8") as f:
    data = yaml.safe_load(f)

print("Loaded sections:")
for key in data.keys():
    print("-", key)
```

## 3. Paper metadata

```python
paper = data["paper"]

print(paper["title"])
print("arXiv:", paper["arxiv"])
print("Version:", paper.get("version", ""))
print("URL:", paper["url"])
print("HTML:", paper.get("html", ""))
print("Year:", paper.get("year", ""))
```

## 4. Seminar context

```python
seminar = data.get("seminar", {})

print("Event:", seminar.get("event", ""))
print("Title:", seminar.get("title", ""))
print("Speaker:", seminar.get("speaker", ""))
print("Affiliation:", seminar.get("affiliation", ""))
print("Date:", seminar.get("date", ""))
print("Context:", seminar.get("context", ""))
```

## 5. Core claim

```python
print(data["core_claim"])
```

## 6. Observations

```python
for i, observation in enumerate(data.get("observations", []), start=1):
    print(f"{i}. {observation}")
```

## 7. Engineering statements

```python
for item in data.get("engineering_statements", []):
    print(f"{item['id']} — {item['category']}")
    print(item["statement"])
    print()
```

## 8. Seminar questions

```python
for i, question in enumerate(data.get("questions_for_seminar", []), start=1):
    print(f"{i}. {question}")
```

## 9. Climate Reality framing

```python
climate = data.get("climate_reality", {})

print(climate.get("title", ""))
print()
print(climate.get("statement", ""))
print()
print("Mathematical context:", climate.get("mathematical_context", ""))
```

## 10. Toward phonon-mediated quantum gates

This section keeps the repo pointed toward the likely next step: phonon-mediated quantum gates.

For now, this is a research roadmap, not a simulation claim.

```python
gate_note = '''
Toward phonon-mediated quantum gates

The immediate paper demonstrates all-mechanical coherence protection and fast
single-qubit control in a silicon-vacancy spin system.

The next research question is whether mechanically dressed protection can remain
compatible with high-fidelity phonon-mediated two-qubit gates.

Working constraints to track:
- spin-phonon coupling strength
- mechanical quality factor
- drive stability
- residual magnetic noise
- heating
- cavity design
- optical initialization and readout
- two-qubit gate fidelity
- integration into on-chip phononic networks
'''

print(gate_note)
```

## 11. Generate `reports/context_summary.md`

This writes a compact Markdown summary from `paper.yaml`.

```python
reports_dir = ROOT / "reports"
reports_dir.mkdir(exist_ok=True)

summary = f'''# {data["climate_reality"]["title"]}

**Paper:** {paper["title"]}  
**arXiv:** {paper["arxiv"]}  
**URL:** {paper["url"]}  

**Seminar:** {seminar.get("title", "")}  
**Speaker:** {seminar.get("speaker", "")}  
**Date:** {seminar.get("date", "")}

## Core claim

{data["core_claim"]}

## Observations

''' + "\n".join(
    f"- {obs}" for obs in data.get("observations", [])
) + '''

## Engineering statements

''' + "\n".join(
    f"- **{item['id']} — {item['category']}:** {item['statement']}"
    for item in data.get("engineering_statements", [])
) + f'''

## Toward phonon-mediated quantum gates

The immediate result supports coherence protection and fast single-qubit control.
The next research direction is to track whether mechanically dressed protection
can remain compatible with high-fidelity phonon-mediated two-qubit gates.

## Climate reality

{data["climate_reality"]["statement"]}

{data["climate_reality"].get("mathematical_context", "")}
'''

out = reports_dir / "context_summary.md"
out.write_text(summary, encoding="utf-8")

print("Wrote:", out)
```
